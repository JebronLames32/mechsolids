```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1001
x = np.linspace(0,6,N)

V = np.piecewise(x,
                [x<3, x>3],
                [lambda x: -2/3*x**2-4, lambda x: 24-4*x])

M = np.piecewise(x,
                [x<3, x>=3],
                [lambda x: -2/9*x**3-4*x, lambda x: -2*(6-x)**2])
```

```python
%matplotlib inline

from matplotlib import rcParams

rcParams['font.size']=14
rcParams['mathtext.fontset']='cm'
```

```python
#fig, axes = plt.subplots(1,2,tight_layout=True)

fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([1.2,0.1,0.8,0.8])

#plt.subplot(1,2,1)
ax1.plot(x,V,'blue')
ax1.set_xlabel(r'$x$',fontsize=20)
ax1.set_ylabel(r'$V$',fontsize=20)
ax1.axhline(y=0,color='k')
ax1.axvline(x=0,color='k')
ax1.fill_between(x,V,color='blue',alpha=0.2)
ax1.set_ylim([-10,12])

ax2.plot(x,M)
ax2.set_xlabel(r'$x$',fontsize=20)
ax2.set_ylabel(r'$M$',fontsize=20)
ax2.axhline(y=0,color='k')
ax2.axvline(x=0,color='k')
ax2.fill_between(x,M,color='blue',alpha=0.2)
#ax2.set_ylim([-18,1])
```
