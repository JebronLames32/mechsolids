```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Mohr2D(sigma_xx, sigma_yy, sigma_xy, axes):
    sigma_avg = (sigma_xx + sigma_yy)/2
    R = np.sqrt( ((sigma_xx - sigma_yy)/2)**2 + sigma_xy**2)
    angle = np.linspace( 0 , 2 * np.pi , 150 ) 
    x = sigma_avg + R*np.cos(angle)
    y = R*np.sin(angle)
    axes.plot(x,y)
    axes.set_aspect(1)
    #return [sigma_avg, R]
```

```python
figure, axes = plt.subplots(1)
axes.axhline(y=0, color='k')
axes.axvline(x=0, color='k')
Mohr2D(30, 5, 10, axes)
```

```python
def Mohr3D(stress_matrix, axes):
    evalues, evectors = np.linalg.eig(stress_matrix)
    evalues = np.sort(evalues)[::-1]
    sigma_p1 = evalues[0]
    sigma_p2 = evalues[1]
    sigma_p3 = evalues[2]
    Mohr2D(sigma_p1, sigma_p2,0,axes)
    Mohr2D(sigma_p2, sigma_p3,0,axes)
    Mohr2D(sigma_p1, sigma_p3,0,axes)
    return [sigma_p1, sigma_p2, sigma_p3]
```

```python
figure, axes = plt.subplots(1)
axes.axhline(y=0,color='k')
axes.axvline(x=0,color='k')
stress_matrix = np.matrix([[15,-3,0],[-3,-4,1],[0,1,10]])
Mohr3D(stress_matrix, axes)
```
